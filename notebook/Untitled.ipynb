```python
from PIL import Image
import os
```

```python
    input_dir = 'chessboards'  # Directory with chessboard images
output_dir = 'chessDataset'
os.makedirs(output_dir, exist_ok=True)
```

```python
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Adjust extensions as needed
        # Load the chessboard image
        image_path = os.path.join(input_dir, filename)
        chessboard = Image.open(image_path)
        
        # Get image dimensions
        width, height = chessboard.size
        square_size = width // 8  # Assuming each board is evenly divided into 8x8 squares

        # Loop over each row and column to crop and save each square
        for row in range(8):
            for col in range(8):
                # Calculate the coordinates of each square
                left = col * square_size
                upper = row * square_size
                right = left + square_size
                lower = upper + square_size
                
                # Crop the square
                square = chessboard.crop((left, upper, right, lower))
                
                # Save each square with a unique name based on the board and position
                square_filename = f"{output_dir}/{os.path.splitext(filename)[0]}{row}_{col}.png"
                square.save(square_filename)
                print(f"Saved {square_filename}")
```
